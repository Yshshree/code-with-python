```python
#Dictionary Methods
#Key()-Returns all keys in the dictonary.
E={'mom':50,'dad':56,'sis':23,'me':20}
print(E.keys())
print(type(E.keys()))
```

```python
#Values() - Returns all values in the dict
E.values()
print(E.values())
print(type(E.values()))
```

```python
#get()-Returns the value for a key(avoids error if key not found)
E.get('sis')
print(E.get('sis'))
```

```python
E.get('uncle','value not exist')
```

```python
E.get('mom','value not exist')
print(E.get('mom'))
E.get('mom',50)
print(E.get('mom'))
E.get('uncle',51)
```

```python
#Update- Updates dictionary with another dictionary or key-value pairs.
E.update({'dad':59})
print(E)
print(id(E))
E.update({'me':24})
print(E)
print(id(E))
```

```python
E.update({'uncle':52})
print(E)
```

```python
F={'aunty':44,'c_bro':11}
E.update(F)
print(E)
print(id(E))
```

```python
#setdefault()-Returns value of the key;inserts key with default value if not present.
E.setdefault('mom',35)
print(E)
E.setdefault('bro',45)
print(E)
```

```python
#pop-Removes and returns the value of a specified key.
E.pop('mom')
print(E)
```

```python
E.clear()
print(E)
```

```python
#SET METHODS
#Union-Returns union of sets.
a={1,2,3,4}
b={3,4,5,6}
a.union(b)
print(a.union(b))
```

```python
#difference-Returns elements in first set but not in second.
a.difference(b)
print(a.difference(b))
```

```python
#intersection - Returns common elements.
a.intersection(b)
print(a.intersection(b))
```

```python
b.difference(a)
print(b.difference(a))
```

```python
a.add(5)
print(a)
a.add(7,8)
print(a)
```

```python
a={1,2,3}
a.update([1,5,7])
print(a)
a.update('YASH')
print(a)
a.update([111,222])
print(a)
a.update({'a':1000,'b':2000})
print(a)
```

```python
a.remove('a')
print(a)
```

```python
#Discard-Removes an element.
x={1,4,5,9,7,3,2}
y={111,333,666}
x.discard('y')
print(x)
```

```python
#Clear-Removes all element
x.clear()
print(x)
```
