```python
#List Methods

x=[1,2,3]
print(type(x))
print(id(x))
```

```python
#Append-It adds a single item to end of a list.
x=[1,2,3]
print(type(x))
print(id(x))
x.append(4)
print(x)
print(id(x))
x.append(5)
print(x)
x.append(6)
print(x)
x.append(7)
print(x)
x.append([5,6,7])
print(x)
```

```python
# Extend-It adds a multiple items from another iterable(list,tuple,set,string)at the end of the list.
x.extend([5,6,7])
print(x)
print(id(x))
x.extend('yash')
print(x)
x.extend({1,3,5})
print(x)

```

```python
#Insert- adds element of another iterable to the list
x.insert(2,"Three")
print(x)
```

```python
x.insert(15,"a")
print(x)
```

```python
#Pop-Removes and returns an element at a given index(default:last)
x=[1,2,3,4,5]
x.pop(4)
print(x)
```

```python
#Remove -Remove element by value
x.remove(2)
print(x)
```

```python
#Sorts-sorts the list in ascending order(in-place).
z=[6,9,5,3,2,1,3]
print(id(x))
z.sort()
print(z)
```

```python
#Reverse- Reverses the order of the list.
y=[1,9,2,6]
y.reverse()
print(y)
```

```python
m=[1,9,6,3,5,2]
m.sort(reverse=True)
print(m)
```

```python
#Clear - removes all elements from the list.
y=[1,2,3,4,5]
y.clear()
print()
```

```python
#Copy - Returns a shallow copy of the list.
u=[1,2,5,4,3]
r=u.copy()
print(u)
print(r)
```
