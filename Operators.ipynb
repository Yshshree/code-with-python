```python
#Arithemetic operators(+,-,*,/,//,%)
a=5
b=10
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a%b)
```

```python
#Comparison operator(<,>,<=,>=,!=,==)
x=10
y=20
print(x>y)
print(x<y)
print(x>=y)
print(x<=y)
print(x!=y)
print(x==y)
```

```python
#logical operator(OR,AND,NOT)
a=10
b=20
print(a<=b)and(a==b)
print((3-2)and(3-3))
print((3-2)or(3-3))
print((25-2)or(3-3))
print((2-25)or(3-3))
print(a>b)and not(a<b)
print(2+3)and not(a>b)


```

```python
#Membership operator(in,not in)
a=(1,2,3)
print(2 in a)
print(0 in a)
print(0 not in a)
print(6 in a)
print(6 not in a)
print((3-2)in a)
```

```python
#Identity operator(is,is not)
a=(1,2,3)
print((a is 1))
print((a is 1))
print((a is not 1))
print((a is (1,2,3)))
print((1,2,3)is a)
```
