```python
#Concatination-It is nothing but the similar type of addition(joining two or more sequence together using "+"operator).
#Examples of Concatination given below-
```

```python
################ STRING ####################

a='Yash'
b='Patil'
c=a+b
print(c)
```

```python
a='Yash'
b='Patil'
c=a+' '+b
print(c)
```

```python
################## Tuple ####################
x=(1,2)
y=(3,4)
z=x+y
print(z)
```

```python
################ List ####################
q=[1,2,3]
r=[4,5,6]
s=q+r
print(s)
```

```python
#Replication- Repetedly values(It means Repeating a sequence multiple times).
#Examples of Replication given below-
```

```python
############# String ############
l='Yash'
m=(l+' ')*3
print(m)
```

```python
################## Tuple ############
x=(2,"Yash")
y=x*3
print(y)
```

```python
#################### List ################
p=[1,2,'yash']
q=(p+[3,4])*2
r=q*2
print(r)
```

```python
#Indexing- Positive and Neagative values(It means Accessing a specific element from a sequences using it position).
#Examples of Indexing given below-
```

```python
############# String ################
x='I am yash'
y=x[3]+x[-4]
print(y)
```

```python
############# Tuple ####################
r=(1,2,11,45,'Yash',56)
q=r[3],r[2],r[4]
print(q)
```

```python
############ List #############
m=['Raj','saniya','soham']
n=m[0][0]+m[1][1]+m[0][2]+m[1][3]+m[1][4]+m[0][1]
o=m[2][2]+m[1][1]+m[1][2]+m[1][3]
p=m[1][0]+m[0][1]+m[1][2]+m[0][1]+m[2][4]
print(n)
print(o)
print(p)
```

```python
m=['Raj','saniya','soham']
n=m[0][0]+m[1][1]+m[0][2]+m[1][3]+m[1][4]+m[0][1],m[2][2]+m[1][1]+m[1][2]+m[1][3],m[1][0]+m[0][1]+m[1][2]+m[0][1]+m[2][4]
print(n)
```

```python
#Slicing -In python is a way to extract a portion(subpart)of a sequence like string,list,tuple using index numbers.
#Examples of slicing given below-
```

```python
############ String ############
a="my name is raj & i love python"
b=a[-13:-7:1]+a[10:14:1]
print(b)
```

```python
c=a[0: :2]
print(c)
```

```python
d=a[ : :-1]
print(d)
```

```python
############### Tuple #############
x=(1,'r',2,'b',3,'k',4,'l')
y=x[2:6:1]
print(y)
```

```python
z=x[-6:-2:1]
print(z)
```

```python
r=x[-4:3:1]
print(r)
```

```python
q=x[-3:-6:-1]
print(q)
```

```python
x=(-1,1,2,-2,3,-3)
m=x[-1:-5:-1]
print(m)
```

```python
############## List #################
m=[1,3,7,59,6,8,11,45]
l=m[-1:-6:-2]
print(l)
```

```python
q=m[6:1:-1]
print(q)
```

```python
g=m[-1:-6:1]
print(g)
```
